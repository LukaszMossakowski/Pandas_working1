## 111. LAB 6: Tabele przestawne

```python
import numpy as np
from datetime import datetime
from datetime import timedelta
import time as time
```

```python
df=pd.read_csv("./marathon_results_2016.csv", usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],
               index_col="Bib")
```

```python
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)

df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
```

```python
df.head()
```

```python
df.pivot_table(index="Age", columns="M/F", values="TotalSeconds", aggfunc="mean").head()
```

```python
dfPivot=df.pivot_table(index="Age", columns="M/F", values=["TotalSeconds", "HalfSeconds"], aggfunc="mean")
dfPivot.head()
```

```python
dfPivot.swaplevel(axis=1).sort_index(axis=1).head()
```

```python
df.pivot_table(index="Age", columns=["M/F"], values="TotalSeconds", aggfunc="count").head()
```

```python
df.pivot_table(index="Country", aggfunc="count", values="TotalSeconds").head()
```

```python
df.pivot_table(index="Country", aggfunc="count", values="TotalSeconds").sort_values("TotalSeconds", ascending=True).head()
```

```python
dfPivot2=df.pivot_table(index=["Country", "State", "City"], values="TotalSeconds", aggfunc="mean").sort_values("TotalSeconds",
                                                                                                     ascending=False)
dfPivot2.head()
```

```python
dfPivot2.iloc[:,0]
```

```python
df["meanTotalSeconds"]=dfPivot2
```

```python
df.head()
```

## 113. LAB 7: Polecenie melt

```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('./marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])

df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)

df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)

df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))

df = df.pivot_table(index="Age",columns="M/F",values="TotalSeconds", aggfunc="mean")
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
df.melt(id_vars="Age", value_vars=["F","M"], value_name="TotalSeconds", var_name="Sex").head()
```

```python
df = pd.read_csv('./marathon_results_2016.csv', index_col='Bib',
 usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
          
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
    
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
          
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
          
df = df.pivot_table(index="Age",columns="M/F",values=["HalfSeconds","TotalSeconds"]).head()
                                                      
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.melt(id_vars="Age", value_name="Time").head()
```

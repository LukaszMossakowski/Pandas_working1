```python
frame["Agency2"].str.split(" ", expand=True, n=10)
```

```python
def getComment(row):
    ReporterType=row["Reporter Type"]
    cost=float(row["Total Operating Expenses"].replace("$",""))
    if cost > 200000:
        comment="Klasa A"
    else:
        comment="Klasa B"
    return ReporterType+"/"+comment
```

```python
frame.apply(getComment, axis=1)
```

```python
frame["Comment"]=frame.apply(getComment, axis=1)
```

```python
frame.head(1)
```

## LAB 5: Operacje na kolumnach tekstowych

```python
fortune=pd.read_csv("./Fortune_500_2017.csv", usecols=['Rank','Title','Industry','Hqlocation','Sector'])
```

```python
fortune.head(1)
```

```python
fortune.Sector=fortune.Sector.str.upper()
```

```python
fortune[fortune.Industry.str.lower().str.contains("comp")].head(1)
```

```python
fortune[["City","State"]]=fortune["Hqlocation"].str.split(",", expand=True)
```

```python
fortune.head(1)
```

```python
def BuildShortcut(row):
    industry=row["Industry"]
    result=""
    for i in industry.split(" "):
        result+=i[0]
    return result
```

```python
BuildShortcut({'Industry':'Factory Under Newspaper'})
```

```python
fortune["IndustryShort"]=fortune.apply(BuildShortcut, axis=1)
```

```python
fortune.head(1)
```

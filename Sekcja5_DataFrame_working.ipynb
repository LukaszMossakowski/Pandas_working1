## LAB 12: Metody isin, isnull, notnull, between

```python
fuel=pd.read_csv("./fuel.csv", usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)',
                                        'Start Stop Technology'], low_memory=False)
```

```python
listOfCars=['Renault', 'Toyota','Ford']
isInList=fuel.Make.isin(listOfCars)
isInList.head()
```

```python
fuel[isInList].head()
```

```python
isNoStartStopDefined=fuel["Start Stop Technology"].isnull()
isNoStartStopDefined.head()
```

```python
fuel[isNoStartStopDefined].head()
```

```python
isStartStopDefined=fuel["Start Stop Technology"].notnull()
```

```python
fuel[isStartStopDefined].head()
```

```python
mpg50_60=fuel["Combined MPG (FT1)"].between(50,60)
```

```python
fuel[mpg50_60].head()
```

```python
print("There are", len(isInList), "cars within the list ['Renault', 'Toyota', 'Ford'].")
```
